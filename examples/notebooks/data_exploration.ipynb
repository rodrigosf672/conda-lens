# Data Exploration
Explore the active Python environment and installed packages using conda-lens.

```python
from pathlib import Path
import sys
sys.path.append(str(Path('src').resolve()))
from conda_lens.env_inspect import get_active_env_info
env = get_active_env_info()
env.name, env.python_version, env.os_info, env.platform_machine
```

```python
total = len(env.packages)
pip_count = sum(1 for p in env.packages.values() if p.manager == 'pip')
conda_count = sum(1 for p in env.packages.values() if p.manager == 'conda')
{"total": total, "pip": pip_count, "conda": conda_count}
```

```python
sorted_pkgs = sorted(env.packages.values(), key=lambda p: p.name)
[(p.name, p.version, p.manager) for p in sorted_pkgs[:25]]
```
